```python
#check the GCR of isothermal atmosphere


from scipy import integrate
import scipy.integrate as integrate
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

SMALL_SIZE = 13
matplotlib.rc('font',size = SMALL_SIZE, family = 'Arial')
matplotlib.rc('axes',titlesize = SMALL_SIZE)

hfont = {'fontname':'Helvetica'}

M_Earth = 5.972e+27
mu = 2.34
m_p = 1.67*10**(-24) #[g]
k_B = 1.38*10**(-16)#[erg/K]
G = 6.67408 * 10**(-8) #[dyne cm^2/g^2]
M_mars = 0.64171*10**(27) #[g]
M_sun = 1.989*10**(33) #[g]
M_earth = 5.98*10**27 #[g]
R_E = 6.378*10**8 #[cm]
autocm = 1.496*10**(13)
Ne_reserve = 64.5e15
S_Ne = 2.7e-12
x_Ne = 2.1e-5

T_0 = 1000

#define temperature
def T_disk(a):
    return T_0*(a/0.1)**(-3/7) #[K]

def Cs_disk(T):
    return np.sqrt(k_B*T/mu/m_p) #CHECK UNITS [cm/s]

def M_int(x, a, M, P_bot,T): 
    #print(Cs_disk(T))
    #compute the R_out based on Boundy/Hill Radisu
    R_H = a*(M/3/M_sun)**(1/3)*autocm #cm
    R_B = G*M/Cs_disk(T)**2 #cm
    

    R_out = [0]
    R_out = min(R_H,R_B)
    R_out_E = R_out/R_E
    
    A = G*M/Cs_disk(T)**2 #CHECK UNITS [cm]

    #P_bot = 10**6#Ne_reserve/S_Ne/x_Ne/M
    R_bot = R_E*M**0.25
    exp = np.exp(-A/R_out + A/R_bot)
    rho_bot = P_bot/Cs_disk(T)**2 #[g/cm^3]
    rho_ne = rho_bot/exp
    
    return 4*np.pi*rho_ne*(x**2*np.exp(A/R_bot - A/x))

M_array = np.linspace(1e-1, 10,5000)
M_P_Constant = [0]*len(M_array)
M_P_Ne_limit = [0]*len(M_array)
M_P_Constant_30 = [0]*len(M_array)
M_P_Ne_limit_30 = [0]*len(M_array)
M_P_Constant_20 = [0]*len(M_array)
M_P_Ne_limit_20 = [0]*len(M_array)
M_P_Constant_50 = [0]*len(M_array)
M_P_Ne_limit_50 = [0]*len(M_array)
M_P_10bar = [0]*len(M_array)
M_P_5bar = [0]*len(M_array)

def r_max(M, T):
    return G*M*M_Earth/Cs_disk(T)**2

def GCR_cal(M, T):
    r_min = R_E*(M)**0.25
    P = Ne_reserve/S_Ne/x_Ne/M/M_Earth
    Mp = quad(M_int, r_min, r_max(M,1500), args=(1.,M*M_Earth, P,T))[0]
    return Mp/M/M_Earth

Mars = GCR_cal(0.5, 1500)
print(Mars)





def GCR_isothermal(a, M):
    T = T_disk(a) 
    r_min = R_E*(M)**0.25
    P_Ne = Ne_reserve/S_Ne/x_Ne/M/M_Earth
    res = quad(M_int, r_min, r_max(M,T), args=(a ,M*M_Earth, P_Ne,T))[0]
    return res/M/M_Earth

N = 100
sma = np.logspace(-2., 1., N)
M_co = np.logspace(-3, 2, N)

fig = plt.figure()
ax1 = fig.add_subplot(111,label = "1",)
i = 0; j = 0;
ta = np.zeros((N,N))

while i < N  :
    j = 0
    while j < N :
        ta[j][i] = GCR_isothermal(sma[i], M_co[j])
        j += 1
    i += 1
    
print(ta)
plt.rcParams["font.family"] = "Arial"

ax.matshow(ta, cmap=plt.cm.Blues)
ax.set_yscale('log')
ax.set_xscale('log')



for i in range(len(M_array)):
    r_min = R_E*(M_array[i])**0.25
    P_Ne = Ne_reserve/S_Ne/x_Ne/M_array[i]/M_Earth
    M_P_Constant[i] = quad(M_int, r_min, r_max(M_array[i],1500), args=(1.,M_array[i]*M_Earth, 10**6,1500))[0]
    M_P_Ne_limit[i] = quad(M_int, r_min, r_max(M_array[i],1500), args=(1.,M_array[i]*M_Earth,P_Ne,1500))[0]
    M_P_Constant_30[i] = quad(M_int, r_min, r_max(M_array[i],3000), args=(1.,M_array[i]*M_Earth, 10**6, 3000))[0]
    M_P_Ne_limit_30[i] = quad(M_int, r_min, r_max(M_array[i],3000), args=(1.,M_array[i]*M_Earth,P_Ne, 3000))[0]
    M_P_Constant_20[i] = quad(M_int, r_min, r_max(M_array[i],2300), args=(1.,M_array[i]*M_Earth, 10**6, 2300))[0]
    M_P_Ne_limit_20[i] = quad(M_int, r_min, r_max(M_array[i],2300), args=(1.,M_array[i]*M_Earth,P_Ne, 2300))[0]
    M_P_Constant_50[i] = quad(M_int, r_min, r_max(M_array[i],5000), args=(1.,M_array[i]*M_Earth, 10**6, 5000))[0]
    M_P_Ne_limit_50[i] = quad(M_int, r_min, r_max(M_array[i],5000), args=(1.,M_array[i]*M_Earth,P_Ne, 5000))[0]
    M_P_10bar[i]  = quad(M_int, r_min, r_max(M_array[i],5000), args=(1.,M_array[i]*M_Earth, 10**7, 5000))[0]
    M_P_5bar[i]  = quad(M_int, r_min, r_max(M_array[i],5000), args=(1.,M_array[i]*M_Earth, 5*10**6, 5000))[0]
fig, ax = plt.subplots(figsize = (8, 6), dpi = 80)    
    
plt.loglog(M_array, np.asarray(M_P_Constant)/M_array/M_Earth, label = 'Isothermal Condition, P = 1bar', color = 'b')
plt.loglog(M_array, np.asarray(M_P_5bar)/M_array/M_Earth, label = 'Isothermal Condition, P = 5bar', color = 'cyan')
plt.loglog(M_array, np.asarray(M_P_10bar)/M_array/M_Earth, label = 'Isothermal Condition, P = 10bar', color = 'dodgerblue')
plt.loglog(M_array, np.asarray(M_P_Ne_limit)/M_array/M_Earth, label = 'Isothermal Condition, P = Ne Dissolve', color = 'orange')
#plt.loglog(M_array, np.asarray(M_P_Constant_15)/M_array/M_Earth, label = 'Isothermal Condition, P = 1bar, a = 1.5AU')
#plt.loglog(M_array, np.asarray(M_P_Ne_limit_15)/M_array/M_Earth, label = 'Isothermal Condition, P = Ne Dissolve, a = 1.5AU')
#plt.loglog(M_array, 5.817e-08/M_array, label = 'Ne_reservoir Limit', color = 'r')
#plt.loglog(M_array, 5.817e-05/M_array, label = 'Ne_reservoir Limit, 99.9% degassing', color = 'g')

plt.legend()
plt.xlabel('Core Mass (M_Earth)')
plt.ylabel('GCR')
plt.title('Isothermal Atmosphere with T = 1500K')
plt.legend(bbox_to_anchor=(1.15, 1.0), loc='upper left')
plt.savefig("Isothermal Atmosphere with T = 1500K.png",dpi=300, bbox_inches='tight')
plt.show()


#Plotting for other temperatures. 

# fig, ax = plt.subplots(figsize = (8, 6), dpi = 80)  

# plt.loglog(M_array, np.asarray(M_P_Constant_20)/M_array/M_Earth, label = 'Isothermal Condition, P = 1bar', color = 'blue')
# plt.loglog(M_array, np.asarray(M_P_Ne_limit_20)/M_array/M_Earth, label = 'Isothermal Condition, P = Ne Dissolve', )
# plt.loglog(M_array, 5.817e-08/M_array, label = 'Ne_reservoir Limit')
# plt.loglog(M_array, 5.817e-05/M_array, label = 'Ne_reservoir Limit, 99.9% degassing')
# plt.legend()
# plt.xlabel('Core Mass (M_Earth)')
# plt.ylabel('GCR')
# plt.title('Isothermal Atmosphere with T = 2300K')
# plt.savefig("Isothermal Atmosphere with T = 2300K.png",dpi=300, bbox_inches='tight')
# plt.show()

# fig, ax = plt.subplots(figsize = (8, 6), dpi = 80)  

# plt.loglog(M_array, np.asarray(M_P_Constant_30)/M_array/M_Earth, label = 'Isothermal Condition, P = 1bar')
# plt.loglog(M_array, np.asarray(M_P_Ne_limit_30)/M_array/M_Earth, label = 'Isothermal Condition, P = Ne Dissolve')
# plt.loglog(M_array, 5.817e-08/M_array, label = 'Ne_reservoir Limit')
# plt.loglog(M_array, 5.817e-05/M_array, label = 'Ne_reservoir Limit, 99.9% degassing')
# plt.legend()
# plt.xlabel('Core Mass (M_Earth)')
# plt.ylabel('GCR')
# plt.title('Isothermal Atmosphere with T = 3000K')
# plt.savefig("Isothermal Atmosphere with T = 3000K.png",dpi=300, bbox_inches='tight')
# plt.show()

# fig, ax = plt.subplots(figsize = (8, 6), dpi = 80)  

# plt.loglog(M_array, np.asarray(M_P_Constant_50)/M_array/M_Earth, label = 'Isothermal Condition, P = 1bar')
# plt.loglog(M_array, np.asarray(M_P_Ne_limit_50)/M_array/M_Earth, label = 'Isothermal Condition, P = Ne Dissolve')
# plt.loglog(M_array, 5.817e-08/M_array, label = 'Ne_reservoir Limit')
# plt.loglog(M_array, 5.817e-05/M_array, label = 'Ne_reservoir Limit, 99.9% degassing')
# plt.legend()
# plt.xlabel('Core Mass (M_Earth)')
# plt.ylabel('GCR')
# plt.title('Isothermal Atmosphere with T = 5000K')
# plt.savefig("Isothermal Atmosphere with T = 5000K.png",dpi=300, bbox_inches='tight')
# plt.show()
```
